```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
```

```python
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
# help(train_test_split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
base_model = DecisionTreeClassifier()
```

```python
base_model.fit(X_train, y_train)
```

```python
y_preds = base_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
print(classification_report(y_test, y_preds))
```

```python
confusion = confusion_matrix(y_test, y_preds)
```

```python
cmd = ConfusionMatrixDisplay(confusion)
cmd.plot()
```

```python
base_model.feature_importances_
```

```python
pd.DataFrame(data = base_model.feature_importances_, index= X.columns, columns=['Feature Importance']).sort_values('Feature Importance')
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(8,8), dpi=150)
plot_tree(base_model, class_names=X.columns, filled=True);
```

```python
def report_model(model):
    model_preds = model.predict(X_test)
    print(classification_report(y_test, model_preds))
    print('\n')
    plt.figure(figsize=(8,8), dpi=150)
    plot_tree(model, feature_names=X.columns, filled=True)
```

```python
report_model(base_model)
```

```python
pruned_model = DecisionTreeClassifier(max_depth=2)
```

```python
pruned_model.fit(X_train, y_train)
```

```python
report_model(pruned_model)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
```

```python
entropy_tree.fit(X_train, y_train)
```

```python
report_model(entropy_tree)
```
