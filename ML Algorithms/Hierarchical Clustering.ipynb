```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/cluster_mpg.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1), dtype=int)
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
plt.figure(figsize=(12,8), dpi=110)
sns.heatmap(scaled_df, cmap='viridis')
```

```python
sns.clustermap(scaled_df)
```

```python
sns.clustermap(scaled_df, row_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
plt.figure(figsize=(12,4), dpi=200)
sns.scatterplot(data=df, x='mpg', y='weight', hue=cluster_labels, palette='viridis')
```
