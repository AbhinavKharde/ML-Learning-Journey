```python
X = df.drop('Country', axis=1) # Because that feature is not much important
```

```python
X.head()
```

```python
X = pd.get_dummies(X, drop_first=True, dtype=int)
```

```python
X.head()
```

```python
sns.heatmap(X.corr())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = [] # Sum of Square Distances

for k in range(2, 30):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)

    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30), ssd, '-o')
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```

```python
# Let's check if we select n_clusters=3

new_model = KMeans(n_clusters=3)
new_model.fit(scaled_X)
```

```python
new_model.labels_
```

```python
X['K=3 KMeans Model'] = new_model.labels_
```

```python
X.corr()['K=3 KMeans Model']
```

```python
# Let's plot this
X.corr()['K=3 KMeans Model'].sort_values().plot(kind='bar')
```
